```python
import pandas as pd
import numpy as np
data_df=pd.read_csv("2021VAERSDATA_clean3.csv", encoding="ISO-8859-1")
vax_df=pd.read_csv("2021VAERSVAX_clean3.csv", encoding="ISO-8859-1")
filtered_data=pd.read_csv("filtered_cols.csv", encoding="ISO-8859-1")
```

```python
df=data_df.merge(vax_df, on="VAERS_ID", how="inner")
df.columns
```

```python
small_df=df[['VAERS_ID', 'STATE', 'AGE_YRS','SEX','DIED', 'L_THREAT', 'ER_VISIT',
       'HOSPITAL','X_STAY', 'DISABLE','VAX_NAME', 'OTHER_MEDS', 'CUR_ILL', 'HISTORY']]
small_df
```

```python
# finding na values in columns
small_df.isna().sum()
```

```python
small_df.dropna(subset=["STATE", "AGE_YRS", "SEX"], inplace=True)
```

```python
small_df["DIED"]=small_df["DIED"].replace(np.nan, 0)
small_df["L_THREAT"]=small_df["L_THREAT"].replace(np.nan, 0)
small_df["ER_VISIT"]=small_df["ER_VISIT"].replace(np.nan, 0)
small_df["HOSPITAL"]=small_df["HOSPITAL"].replace(np.nan, 0)
small_df["X_STAY"]=small_df["X_STAY"].replace(np.nan, 0)
small_df["DISABLE"]=small_df["DISABLE"].replace(np.nan, 0)

small_df=small_df.replace("Y",1)

small_df
```

```python
def risk_score(row):
    if row["DIED"] or row["L_THREAT"] or row["ER_VISIT"] or row["HOSPITAL"] or row["X_STAY"] or row["DISABLE"]:
        return 1
    else:
        return 0

small_df["risk"]=small_df.apply(risk_score, axis=1)
small_df
```

```python
small_df=small_df[(small_df["VAX_NAME"]=="COVID19 (COVID19 (MODERNA))") | (small_df["VAX_NAME"]=="COVID19 (COVID19 (PFIZER-BIONTECH))") | (small_df["VAX_NAME"]=="COVID19 (COVID19 (JANSSEN))")]
# small_df=small_df[(small_df["VAX_NAME"]=="COVID19 (COVID19 (PFIZER-BIONTECH))") | (small_df["VAX_NAME"]=="COVID19 (COVID19 (JANSSEN))")]
small_df["VAX_NAME"].value_counts()
```
